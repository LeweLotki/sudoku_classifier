Notebook to basic data anaylsis. There is data imbalance and average word count displayed. Processed data based on analysis is saved to ../data/data_processed.csv

## Data Imbalance

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_path = '../data/data.csv'
df = pd.read_csv(data_path)

df['target'] = df['target'].replace({1: 1, 2: 1, 3: 2, 4: 3, 5: 3})

# Display class distribution (target column)
class_distribution = df['target'].value_counts()

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(class_distribution, labels=class_distribution.index, autopct='%1.1f%%', colors=['#ff9999','#66b3ff','#99ff99'])
plt.title('Class Imbalance in the Dataset')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()
```

## Average word count by class

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_path = '../data/data.csv'
df = pd.read_csv(data_path)

# Replace values in target column
df['target'] = df['target'].replace({1: 1, 2: 1, 3: 2, 4: 3, 5: 3})

# Calculate word count in features column
df['word_count'] = df['features'].apply(lambda x: len(str(x).split()))

# Group by target class and calculate the average word count
average_word_count = df.groupby('target')['word_count'].mean()

# Plot the bar chart
plt.figure(figsize=(8, 6))
average_word_count.plot(kind='bar', color=['#ff9999','#66b3ff','#99ff99'])
plt.title('Average Word Count by Class')
plt.xlabel('Class')
plt.ylabel('Average Word Count')
plt.xticks(rotation=0)
plt.show()
```

## Distiubtuion of word count

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
data_path = '../data/data.csv'
df = pd.read_csv(data_path)

# Replace values in target column
df['target'] = df['target'].replace({1: 1, 2: 1, 3: 2, 4: 3, 5: 3})

# Calculate word count in features column
df['word_count'] = df['features'].apply(lambda x: len(str(x).split()))

# Plot the distribution of word counts (with x-axis limited to 0-300)
plt.figure(figsize=(10, 6))
plt.hist(df['word_count'], bins=3000, color='#66b3ff', edgecolor='black')
plt.title('Distribution of Word Count in Features Column (0 to 300)')
plt.xlabel('Word Count')
plt.ylabel('Frequency')
plt.xlim(0, 300)  # Limit the x-axis to 0-300
plt.grid(True)
plt.show()
```

## Dropping useless data

```python
import pandas as pd

# Load the data
data_path = '../data/data.csv'
df = pd.read_csv(data_path)

# Replace values in the target column
# df['target'] = df['target'].replace({1: 1, 2: 1, 3: 2, 4: 3, 5: 3})

# Calculate word count in features column
df['word_count'] = df['features'].apply(lambda x: len(str(x).split()))

# Filter the rows where word count is between 10 and 300
df_filtered = df[(df['word_count'] >= 10) & (df['word_count'] <= 300)]

# Drop the 'word_count' column as it's no longer needed
df_filtered = df_filtered.drop(columns=['word_count'])

# Save the processed dataframe to a CSV file
output_path = '../data/data_processed.csv'
df_filtered.to_csv(output_path, index=False)

print(f"Processed data saved to {output_path}")
```

## Word Clouds

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Combine all the features text for the whole dataset
all_text = ' '.join(df['features'])

# Generate word cloud for the whole dataset
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_text)

# Plot the WordCloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for All Puzzle Descriptions')
plt.show()
```

```python
# Create word clouds for each class (easy, medium, hard)
classes = df['target'].unique()

for class_label in classes:
    class_text = ' '.join(df[df['target'] == class_label]['features'])
    
    wordcloud = WordCloud(width=800, height=400, background_color='white').generate(class_text)
    
    # Plot the WordCloud image for each class
    plt.figure(figsize=(10, 5))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for Class {class_label}')
    plt.show()
```

## Unique words

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
df = pd.read_csv('../data/data_processed.csv')

# Function to count unique words
def count_unique_words(text):
    words = text.split()
    unique_words = set(words)
    return len(unique_words)

# Add a new column 'unique_word_count'
df['unique_word_count'] = df['features'].apply(count_unique_words)

# Plot bar chart showing average unique word count for each class
plt.figure(figsize=(10, 6))
sns.barplot(x='target', y='unique_word_count', data=df, estimator='mean', ci=None, palette='Set2')

plt.title('Average Unique Word Count for Each Puzzle Difficulty Class')
plt.xlabel('Puzzle Difficulty Class (1 = Easy, 2 = Medium, 3 = Hard)')
plt.ylabel('Average Unique Word Count')
plt.show()
```
